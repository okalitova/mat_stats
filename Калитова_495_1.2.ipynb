```python
import scipy.stats as sts
import scipy.misc as sm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

–°–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –≤—ã–±–æ—Ä–∫—É ùëã1, ..., ùëãùëÅ –∏–∑ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º ùúÉ = 1 –¥–ª—è ùëÅ = 10^4.

```python
theta = 1.
samples = sts.expon.rvs(size=10000, loc=0, scale=theta)
N = np.size(samples)
```

–î–ª—è –≤—Å–µ—Ö ùëõ <= ùëÅ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –æ—Ü–µ–Ω–∫—É (Ô∏Åùëò!/ùëã^ùëò)Ô∏Å^ (1/ùëò)–ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùúÉ = 1. 

```python
k = np.arange(10) + 1.
res = [[] for x in range(10)] 
#res = np.arange((N + 1) * 10).reshape(10, N + 1)

for i in range(10):
    for n in range(1, N + 1) :
        ans = (sm.factorial(k[i]) / np.mean(samples[:n] ** k[i])) ** (1. / k[i])
        res[i].append(np.abs(np.array(ans - theta)))
```

–ü—Ä–æ–≤–µ–¥–∏—Ç–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ, –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –∑–∞–¥–∞—á–µ, –∏ –≤—ã—è—Å–Ω–∏—Ç–µ, –ø—Ä–∏ –∫–∞–∫–æ–º ùëò –æ—Ü–µ–Ω–∫–∞ –≤–µ–¥–µ—Ç —Å–µ–±—è –ª—É—á—à–µ (—Ä–∞—Å—Å–º–æ—Ç—Ä–∏—Ç–µ –Ω–µ –º–µ–Ω–µ–µ 10 —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π ùëò).

```python
colors = ['red', 'blue', 'green', 'black', 'purple']
x = np.arange(N) + 1.
plt.xlabel("# samples")
plt.ylabel("statistics")
plt.ylim([0., 0.2])
plt.xlim([1., N])
for i in range(5) :
    plt.plot(x, res[i], label="$k = " + str(k[i]) + "$")
plt.legend()
plt.show()
x = np.arange(N) + 1.
plt.xlabel("# samples")
plt.ylabel("statistics")
plt.ylim([0., 0.2])
plt.xlim([1., N])
for i in range(5) :
    plt.plot(x, res[i + 5], label="$k = " + str(k[i + 5]) + "$")
plt.legend()
plt.show()
```
