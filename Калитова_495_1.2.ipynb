```python
import scipy.stats as sts
import scipy.misc as sm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Сгенерируйте выборку 𝑋1, ..., 𝑋𝑁 из экспоненциального распределения с параметром 𝜃 = 1 для 𝑁 = 10^4.

```python
theta = 1.
samples = sts.expon.rvs(size=10000, loc=0, scale=theta)
N = np.size(samples)
```

Для всех 𝑛 <= 𝑁 посчитайте оценку (︁𝑘!/𝑋^𝑘)︁^ (1/𝑘)параметра 𝜃 = 1. 

```python
k = np.arange(10) + 1.
res = [[] for x in range(10)] 
#res = np.arange((N + 1) * 10).reshape(10, N + 1)

for i in range(10):
    for n in range(1, N + 1) :
        ans = (sm.factorial(k[i]) / np.mean(samples[:n] ** k[i])) ** (1. / k[i])
        res[i].append(np.abs(np.array(ans - theta)))
```

Проведите исследование, аналогичное предыдущей задаче, и выясните, при каком 𝑘 оценка ведет себя лучше (рассмотрите не менее 10 различных значений 𝑘).

```python
colors = ['red', 'blue', 'green', 'black', 'purple']
x = np.arange(N) + 1.
plt.xlabel("# samples")
plt.ylabel("statistics")
plt.ylim([0., 0.2])
plt.xlim([1., N])
for i in range(5) :
    plt.plot(x, res[i], label="$k = " + str(k[i]) + "$")
plt.legend()
plt.show()
x = np.arange(N) + 1.
plt.xlabel("# samples")
plt.ylabel("statistics")
plt.ylim([0., 0.2])
plt.xlim([1., N])
for i in range(5) :
    plt.plot(x, res[i + 5], label="$k = " + str(k[i + 5]) + "$")
plt.legend()
plt.show()
```

При всех k при увеличении выборки оценка становится ближе к тета. Но все-таки самая лучшая оценка получается при k = 1.
